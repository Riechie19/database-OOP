```python
import abc 
from abc import ABC
```

```python
class Branch:
    __branchName = None
    __branchID = None
    __branchAddress = None
    __totalEmployee = 0
    
    def __init__(self, name: str, ID: int, address: str):
        self.branchName = name
        self.branchID = ID
        self.branchAddress = address
    
    def addEmployee(self):
        self.totalEmployee += 1
    
    def substractEmployee(self):
        self.totalEmployee -= 1
    
    @property
    def branchID(self):
        return self.__branchID
    
    @branchID.setter
    def branchID(self, value: int):
        self.__branchID = value
    
    @property 
    def branchName(self):
        return self.__branchName
    
    @branchName.setter
    def branchName(self, name: str):
        self.__branchName = name
    
    @property
    def branchAddress(self):
        return self.__branchAddress
    
    @branchAddress.setter
    def branchAddress(self, address: str):
        self.__branchAddress = address
    
    @property
    def totalEmployee(self):
        return self.__totalEmployee
    
    @totalEmployee.setter
    def totalEmployee(self, value: int):
        self.__totalEmployee = value
```

```python
class Employee(ABC):
    _employeeID = None
    _name = None
    _employeeAddress = None
    _birthDate = None
    _gender = None
    _salary = None
    _branch = None
    _position = None
    
    def __init__(self, ID: int, name: str, address: str, birthday: str, gender: str, salary: float, branch: Branch):
        self.employeeID = ID
        self.name = name
        self.employeeAddress = address
        self.birthDate = birthday
        self.gender = gender
        self.salary = salary
        self.branch = branch
    
    @property
    def employeeID(self):
        return self._employeeID
    
    @employeeID.setter
    def employeeID(self, value: int):
        self._employeeID = value
    
    @property
    def name(self):
        return self._name
    
    @name.setter
    def name(self, name: str):
        self._name = name
    
    @property
    def birthDate(self):
        return self._birthDate
    
    @birthDate.setter
    def birthDate(self, birthday: str):
        self._birthDate = birthday
    
    @property
    def gender(self):
        return self._gender
    
    @gender.setter
    def gender(self, gender: str):
        self._gender = gender
    
    @property
    def salary(self):
        return self._salary
    
    @salary.setter
    def salary(self, salary: float):
        self._salary = salary
    
    @property
    def branch(self):
        return self._branch
    
    @branch.setter
    def branch(self, branch: Branch):
        if self.branch != None:
            self.branch.substractEmployee()
        self._branch = branch
        self._branch.addEmployee()
    
    @abc.abstractproperty
    def position(self):
        pass
```

```python
class Sales(Employee):
    __totalSale = 0
    __tenantRating = None
    def __init__(self, ID: int, name: str, address: str, birthday: str, gender: str, salary: float, branch: Branch, tenantRating: float):
        super().__init__(ID, name, address, birthday, gender, salary, branch)
        self.tenantRating = tenantRating
        self.position = 'Sales'
    
    @property
    def tenantRating(self):
        return self.__tenantRating
    
    @tenantRating.setter
    def tenantRating(self, tenantRating: float):
        self.__tenantRating = tenantRating
    
    @property
    def position(self):
        return self._position
    
    @position.setter
    def position(self, position: str):
        self._position = position
    
    @property
    def totalSale(self):
        return self.__totalSale
    
    @totalSale.setter
    def totalSale(self, totalSale: int):
        self.__totalSale = totalSale
    
    def addSale(self):
        self.totalSale += 1
```

```python
class Owner:
    __name = None
    __address = None
    __phoneNumber = None
    __email = None
    __contactPerson = None
    __totalProperty = 0
    
    def __init__(self, name: str, address: str, phoneNumber: int, email: str, contactPerson: Relation):
        self.name = name
        self.address = address
        self.phoneNumber = phoneNumber
        self.email = email
        self.contactPerson = contactPerson
    
    @property
    def name(self):
        return self.__name
    
    @name.setter
    def name(self, name: str):
        self.__name = name
    
    @property
    def address(self):
        return self.__address
    
    @address.setter
    def address(self, address: str):
        self.__address = address
    
    @property
    def phoneNumber(self):
        return self.__phoneNumber

    @phoneNumber.setter
    def phoneNumber(self, phoneNumber: int):
        self.__phoneNumber = phoneNumber
    
    @property
    def email(self):
        return self.__email
    
    @email.setter
    def email(self, email: str):
        self.__email = email
    
    @property
    def contactPerson(self):
        return self.__contactPerson
    
    @contactPerson.setter
    def contactPerson(self, value: Relation):
        if self.__contactPerson!= None:
            self.__contactPerson.substractRelation()
        self.__contactPerson = value
        self.__contactPerson.addRelation()
    
    @property
    def totalProperty(self):
        return self.__totalProperty
    
    @totalProperty.setter
    def totalProperty(self, value: int):
        self.__totalProperty = value
    
    def addTotalProperty(self):
        self.totalProperty += 1
```

```python
class Property(ABC):
    _propertyID = None
    _propertyAddress = None
    _width = None
    _length = None
    _price = None
    _owner = None
    _sales = None
    _sold = False
    _area = None
    _type = None
    
    def __init__(self, propertyID: int, propertyAddress: str, width: float, length: float, price: float, owner: Owner, sales: Sales):
        self.propertyID = propertyID
        self.propertyAddress = propertyAddress
        self.width = width
        self.length = length
        self.price = price
        self.owner = owner
        self.sales = sales
        
    @property
    def propertyID(self):
        return self._propertyID
    
    @propertyID.setter
    def propertyID(self, propertyID: int):
        self._propertyID = propertyID

    @property
    def propertyAddress(self):
        return self._propertyAddress
    
    @propertyAddress.setter
    def propertyAddress(self, propertyAddress: str):
        self._propertyAddress = propertyAddress

    @property
    def width(self):
        return self._width

    @width.setter
    def width(self, width: float):
        self._width = width

    @property
    def length(self):
        return self._length

    @length.setter
    def length(self, length: float):
        self._length = length

    @property
    def price(self):
        return self._price

    @price.setter
    def price(self, price: float):
        self._price = price

    @property
    def owner(self):
        return self._owner

    @owner.setter
    def owner(self, owner: Owner):
        self._owner = owner
        self._owner.addTotalProperty()

    @property
    def sales(self):
        return self._sales

    @sales.setter
    def sales(self, sales: Sales):
        self._sales = sales
    
    @property
    def sold(self):
        return self._sold
    
    @sold.setter
    def sold(self, value: bool):
        self._sold = value
    
    @abc.abstractproperty
    def type(self):
        pass
    
    def getArea(self):
        return self.width * self.length
```

```python
class WorkOrder:
    __workOrderID = None
    __date = None
    __tenant = None
    __sales = None
    __propertyInterested = None
    __status = 'Ongoing'
    
    def __init__(self, workOrderID: int, date: str, tenant: Tenant, sales: Sales, propertyInterested: Property):
        self.workOrderID = workOrderID
        self.date = date
        self.tenant = tenant
        self.sales = sales
        self.propertyInterested = propertyInterested
        self.tenant.addInterested()
    
    @property
    def workOrderID(self):
        return self.__workOrderID
    
    @workOrderID.setter
    def workOrderID(self, ID: int):
        self.__workOrderID = ID
    
    @property
    def date(self):
        return self.__date
    
    @date.setter
    def date(self, date: str):
        self.__date = date

    @property
    def tenant(self):
        return self.__tenant

    @tenant.setter
    def tenant(self, tenant: Tenant):
        self.__tenant = tenant

    @property
    def sales(self):
        return self.__sales

    @sales.setter
    def sales(self, sales: Sales):
        self.__sales = sales

    @property
    def propertyInterested(self):
        return self.__propertyInterested

    @propertyInterested.setter
    def propertyInterested(self, value: Property):
        self.__propertyInterested = value
    
    @property
    def status(self):
        return self.__status
    
    @status.setter
    def status(self, status: str):
        if status == 'Success':
            self.tenant.substractInterested()
            self.sales.addSale()
            self.propertyInterested.sold = True
            self.__status = status
        elif status == 'fail':
            self.tenant.substractInterested()
            self.__status = status
        else:
            print('Work Order final Status must either be success or fail')
```

```python
class Transaction:
    __transactionID = None
    __date = None
    __startDate = None
    __endDate = None
    __workOrder = None
    __admin = None
    __finalPrice = None
    __setupCost = None
    
    def __init__(self, transactionID: int, date: str, startDate: str, endDate: str, workOrder: WorkOrder, admin: Administration, finalPrice: float, setupCost: float):
        self.transactionID = transactionID
        self.date = date
        self.startDate = startDate
        self.endDate = endDate
        self.workOrder = workOrder
        self.admin = admin
        self.finalPrice = finalPrice
        self.setupCost = setupCost
    
    @property
    def transactionID(self):
        return self.__transactionID
    
    @transactionID.setter
    def transactionID(self, ID):
        self.__transactionID = ID
    
    @property
    def date(self):
        return self.__date
    
    @date.setter
    def date(self, date: str):
        self.__date = date

    @property
    def startDate(self):
        return self.__startDate

    @startDate.setter
    def startDate(self, value: str):
        self.__startDate = value

    @property
    def endDate(self):
        return self.__endDate

    @endDate.setter
    def endDate(self, value: str):
        self.__endDate = value

    @property
    def workOrder(self):
        return self.__workOrder

    @workOrder.setter
    def workOrder(self, workOrder: WorkOrder):
        self.__workOrder = workOrder
        self.workOrder.status = 'Success'

    @property
    def admin(self):
        return self.__admin

    @admin.setter
    def admin(self, admin: Administration):
        self.__admin = admin

    @property
    def finalPrice(self):
        return self.__finalPrice

    @finalPrice.setter
    def finalPrice(self, finalPrice: float):
        self.__finalPrice = finalPrice

    @property
    def setupCost(self):
        return self.__setupCost

    @setupCost.setter
    def setupCost(self, setupCost: float):
        self.__setupCost = setupCost
    
    def profit(self):
        return self.finalPrice - self.setupCost
    
```
